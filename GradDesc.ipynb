```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(func, theta, step, iters):
    vals = [None] * (iters+1)
    vals[0] = list(theta)
    
    for i in range(iters):
        for j in range(len(theta)):
            theta[j] = np.array(theta[j]) - step*(func(theta))[j]
        vals[i+1] = list(theta)
        
    return theta, vals
```

```python
def test(x):
    return [2*x[0], 10*x[1]]

theta, vals = gradient_descent(test, [1, 1], 0.1, 100)
```

```python
xs = [x[0] for x in vals]
ys = [x[1] for x in vals]
plt.title("Iterate Pairs")
plt.xlabel("x_1 coordinate")
plt.ylabel("x_2 coordinate")
plt.plot(xs, ys)
```

```python
f = [None] * (len(vals))
x = range(len(vals))

for i in range(len(vals)):
    f[i] = np.linalg.norm(np.array(vals[i]) - np.array([0, 0]))

plt.title("Value vs time")
plt.xlabel("Time")
plt.ylabel("Value")
plt.plot(x, f) 
```

```python
def step(x):
    return 1/(np.sqrt(x))

def gradient_descent_new(func, theta, stepf, iters):
    vals = [None] * (iters+1)
    vals[0] = list(theta)
    
    for i in range(iters):
        for j in range(len(theta)):
            theta[j] = np.array(theta[j]) - stepf(i+1)*(func(theta))[j]
        vals[i+1] = list(theta)
        
    return theta, vals

theta, vals = gradient_descent_new(np.sign, [2], step, 100)

x = range(len(vals))
xs = [x[0] for x in vals]

plt.title("Value vs time")
plt.xlabel("Time")
plt.ylabel("Value")
plt.plot(x, xs)
```
